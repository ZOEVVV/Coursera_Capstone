```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://s3.us.cloud-object-storage.appdomain.cloud/cf-courses-data/CognitiveClass/DP0701EN/version-2/Data-Collisions.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df1=df[['SEVERITYCODE', 'ADDRTYPE', 'INATTENTIONIND', 'UNDERINFL', 'WEATHER', 'ROADCOND', 'LIGHTCOND', 'SPEEDING']]
```

```python
df2=df1.fillna(0)
```

```python
df2['SPEEDING'].replace('Y', 1, inplace=True)
```

```python
df2['INATTENTIONIND'].replace('Y', 1, inplace=True)
```

```python
df2['UNDERINFL'].replace('Y', 1, inplace=True)
```

```python
df2['UNDERINFL'].replace('N', 0, inplace=True)
```

```python
df2['ADDRTYPE'].replace('Intersection', 1, inplace=True)
df2['ADDRTYPE'].replace('Block', 2, inplace=True)
df2['ADDRTYPE'].replace('Alley', 3, inplace=True)
```

```python
df2['WEATHER'].replace('Clear', 1, inplace=True)
df2['WEATHER'].replace('Raining', 2, inplace=True)
df2['WEATHER'].replace('Overcast', 3, inplace=True)
df2['WEATHER'].replace('Unknown', 0, inplace=True)
df2['WEATHER'].replace('Snowing', 4, inplace=True)
df2['WEATHER'].replace('Fog/Smog/Smoke', 4, inplace=True)
df2['WEATHER'].replace('Sleet/Hail/Freezing Rain', 4, inplace=True)
df2['WEATHER'].replace('Blowing Sand/Dirt', 4, inplace=True)
df2['WEATHER'].replace('Severe Crosswind', 4, inplace=True)
df2['WEATHER'].replace('Partly Cloudy', 3, inplace=True)
df2['WEATHER'].replace('Other', 0, inplace=True)
```

```python
df2['ROADCOND'].replace('Dry', 1, inplace=True)
df2['ROADCOND'].replace('Wet', 2, inplace=True)
df2['ROADCOND'].replace('Unknown', 0, inplace=True)
df2['ROADCOND'].replace('Partly Cloudy', 0, inplace=True)
df2['ROADCOND'].replace('Ice', 3, inplace=True)
df2['ROADCOND'].replace('Snow/Slush', 3, inplace=True)
df2['ROADCOND'].replace('Other', 0, inplace=True)
df2['ROADCOND'].replace('Standing Water', 3, inplace=True)
df2['ROADCOND'].replace('Sand/Mud/Dirt', 3, inplace=True)
df2['ROADCOND'].replace('Oil', 3, inplace=True)
```

```python
df2['LIGHTCOND'].replace('Daylight', 1, inplace=True)
df2['LIGHTCOND'].replace('Dark - Street Lights On', 2, inplace=True)
df2['LIGHTCOND'].replace('Unknown', 0, inplace=True)
df2['LIGHTCOND'].replace('Dusk', 3, inplace=True)
df2['LIGHTCOND'].replace('Dawn', 3, inplace=True)
df2['LIGHTCOND'].replace('Dark - No Street Lights', 4, inplace=True)
df2['LIGHTCOND'].replace('Dark - Street Lights Off', 4, inplace=True)
df2['LIGHTCOND'].replace('Other', 0, inplace=True)
df2['LIGHTCOND'].replace('Dark - Unknown Lighting', 0, inplace=True)
```

```python
X=df2[['ADDRTYPE','INATTENTIONIND', 'UNDERINFL', 'WEATHER', 'ROADCOND', 'LIGHTCOND', 'SPEEDING']]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y=df2[['SEVERITYCODE']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
Ks=10
mean_acc=np.zeros((Ks-1))
```

```python
for n in range (1, Ks):
    neigh=KNeighborsClassifier(n_neighbors=n).fit(X_train, np.ravel(y_train,order='C'))
    yhat=neigh.predict(X_test)
    mean_acc[n-1]=metrics.accuracy_score(y_test, yhat)
    
mean_acc.max(), mean_acc.argmax()+1
```

```python
neigh2=KNeighborsClassifier(n_neighbors=8).fit(X_train, np.ravel(y_train,order='C'))
yhat2=neigh2.predict(X_test)
   
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import jaccard_similarity_score
f1_score(y_test, yhat2), jaccard_similarity_score(y_test, yhat2)
```
